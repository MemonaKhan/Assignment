```python
# Question.5

class Car():
    def __init__(self,model,color,name,year,speed):
        self.model = model
        self.color = color
        self.name = name
        self.year = year
        self.speed = speed
            
    def update_color(self,new_color):
        self.color = new_color
    def accelerate(self):
        self.speed +=5
    def brake(self):
        self.speed -=5
        
car1 = Car('BR V','Red','Honda',2018,220)
print("Car 1 Info")
print("Model = ", car1.model)
print("Color = ", car1.color)
print("Name = ", car1.name)
print("Year = ", car1.year)
print("Speed = ", car1.speed)
car1.update_color('Black')
print("After update Color = ", car1.color)
print()

car2 = Car('Corolla','Blue','Toyota',2019,240)
print("Car 2 Info")
print("Model = ", car2.model)
print("Color = ", car2.color)
print("Name = ", car2.name)
print("Year = ", car2.year)
print("Speed = ", car2.speed)
car2.accelerate()
print("After accelerate Speed = ",car2.speed)
print()

car3 = Car('City','Yellow','Honda',2018,210)
print("Car 3 Info")
print("Model = ", car3.model)
print("Color = ", car3.color)
print("Name = ", car3.name)
print("Year = ", car3.year)
print("Speed = ", car3.speed)
car3.brake()
print("After brake speed = ",car3.speed)
print()

car4 = Car('Benz C Class','red','Mercedes',2017,260)
print("Car 4 Info")
print("Model = ", car4.model)
print("Color = ", car4.color)
print("Name = ", car4.name)
print("Year = ", car4.year)
print("Speed = ", car4.speed)
print()

car5 = Car('A4','Pink','Audi',2019,300)
print("Car 5 Info")
print("Model = ", car5.model)
print("Color = ", car5.color)
print("Name = ", car5.name)
print("Year = ", car5.year)
print("Speed = ", car5.speed)
```
